```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import skextremes as ske
```

```python
df = pd.read_csv("Data\data.csv")

shots = df.groupby('game_id')['shot type'].count()
goals = df.loc[df['shot type'] == 'Goal']
goals = goals.groupby('game_id')['shot type'].count()
```

## Total shots

**Fr√©chet / Weibull**

```python
model_shots_GEV = ske.models.classic.GEV(shots.to_numpy(dtype='float32'), fit_method = 'mle', ci = 0.05,
                              ci_method = 'delta')
```

```python
print(model_shots_GEV.params)
model_shots_GEV.plot_summary()
plt.show()
```

**Gumbel**

```python
model_shots_gumbel = ske.models.classic.Gumbel(shots.to_numpy(dtype='float32'), fit_method = 'mle', ci = 0.05,
                              ci_method = 'delta')
```

```python
print(model_shots_gumbel.params)
model_shots_gumbel.plot_summary()
plt.show()
```
